```python
#11-01-2025
import pandas as pd

# Create the dataset
data = {
    "Name": ["John", "Alice", "Bob", "Emma"],
    "Department": ["IT", "HR", "Marketing","IT"],
    "Salary": [60000, 75000, 50000, 65000]
}
df = pd.DataFrame(data)

# Use map to add a prefix to the Department names
df["Name"] = df["Name"].map(lambda x: "Shri " + x)

df 
```

```python
df["Salary"] = df["Salary"].map(lambda x: 100 + x)
```

```python
df
```

```python
df["Salary"]=df["Salary"].apply(lambda x:x*1.10)
```

```python
df
```

```python
#apply row-wise to combine "Name" and "department"
df["employee_info"]=df.apply(lambda row : f"{row["Name"]} ({row["Department"]})",axis=1)
```

```python
df
```

```python
# Low: Salary < 60000
# Medium: 60000 >= Salary <= 70000
# High: Salary  70000

def categorize_salary(row):
    if row["Salary"] < 60000:
        return "Low"
    elif ( row["Salary"] >= 60000 ) and ( row["Salary"] <= 70000 ):
        return "Medium"
    else:
        return "High"

df["Salary_Category"] = df.apply(categorize_salary, axis=1)
df 
```

```python
####Pivot table

data = {
    "Name": ["John", "Alice", "Bob", "Emma", "Sophia", "James"],
    "Department": ["IT", "HR", "Marketing", "IT", "HR", "Marketing"],
    "Salary": [60000, 75000, 50000, 65000, 72000, 55000],
    "Bonus": [5000, 7000, 3000, 6000, 6500, 3500]
}
df = pd.DataFrame(data)
df 
```

```python
pivot = pd.pivot_table(
    df,
    values=["Salary", "Bonus"],     
    index="Department",            
    aggfunc={"Salary": "mean", "Bonus": "sum"}  # Aggregation functions
)
pivot
```
